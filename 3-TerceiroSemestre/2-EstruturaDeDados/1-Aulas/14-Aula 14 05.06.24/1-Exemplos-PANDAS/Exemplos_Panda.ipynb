```python
#Importando biblioteca panda
import pandas as pd

#Importando uma planilha do Execel
#Necessário fazer Upload do arquivo
vendas_pd = pd.read_excel("/baseDados.xlsx")

#Mostrando uma linha selecionada
print("Imprimindo dados da linha 0:")
display(vendas_pd.loc[0])

#Mostrando dados de mais de uma linha selecionada
print("\n\nImprimindo dados da linha 0 e 1")
display(vendas_pd.loc[0:1])


#Mostrando dados ENTRE linha 0 e linha 3
print("\n\nImprimindo dados entre linha 0 e linha 3")
display(vendas_pd.loc[0:3])


#Mostrando dados da linha 0 e linha 3 SOMENTE
print("\n\nImprimindo dados da linha 0 e linha 3")
display(vendas_pd.loc[[0,3]])
```
